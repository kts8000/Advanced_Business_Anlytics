```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'PayDate': pd.date_range(start='2016-01-01', periods=200, freq='D'),
    'Dollars': [x * 100 for x in range(200)]  # Dollars 열에 예시 값
}
df = pd.DataFrame(data)

filtered_df = df[(df['PayDate'] >= '2016-03-01') & (df['PayDate'] <= '2016-07-31')]

filtered_df['Month'] = filtered_df['PayDate'].dt.to_period('M')

monthly_sum = filtered_df.groupby('Month')['Dollars'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sum['Month'].astype(str), monthly_sum['Dollars'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Month (2016)')
plt.ylabel('Total Dollars')
plt.title('Total Dollars by Month (March 2016 - July 2016)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
summary_vendor = df1.groupby('VendorNumber').agg({
                'LeadTime': ['mean', 'min', 'max'],
                'OrderTime': ['mean', 'min', 'max'],
                'PaymentLT': ['mean',  'min', 'max']
            })
```

```python
# Get the sorted top 5 rows based on 'OrderTime' mean
top_bad_5 = summary_vendor.sort_values(('OrderTime', 'mean'), ascending=False).head(5)
top_bad_10 = summary_vendor.sort_values(('OrderTime', 'mean'), ascending=False).head(10)

# Retrieve the index of the top 5 rows
top_bad_5_index = top_bad_5.index.tolist()
top_bad_10_index = top_bad_10.index.tolist()

print(top_bad_5_index)
print(top_bad_10_index)
```

```python
import pandas as pd
import plotly.express as px

filtered_df = df1[df1['InventoryCategory'].isin(['MOUNTMEND', 'GOULCREST', 'DONCASTER'])]

podate_att = filtered_df.groupby(['PODate', 'InventoryCategory']).apply(lambda x: x['PODate'].count()).reset_index(name='Counts')

fig = px.line(podate_att, x='PODate', y='Counts', color='InventoryCategory', 
              title='Purchase Order Counts by Date and InventoryCategory (Filtered)')

fig.show()
```
